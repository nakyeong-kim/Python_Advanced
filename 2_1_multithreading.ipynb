# 1. Multithreading

### 1) Difference between Process and Thread<br><br>

image.png
<br><br>

* Process
  1. 운영체제에서 할당 받는 자원의 단위를 의미 (= 실행 중인 프로그램)
  2. CPU 동작 시간과 주소 공간이 모두 독립적 ex) 크롬 실행하면서 익스플로어 실행
  3. Code, Data, Stack, Heap 영역 모두 독립적
  4. 최소 1개의 메인 Thread 보유 ex) 크롬 창 1개가 process, 여러 탭이 thread
  5. 파이프, 파일, 소켓 등을 사용해서 프로세스 간 통신 가능 = Context Switching → Cost 높음

* Thread
  1. Process 내의 실행 흐름 단위
  2. Process의 자원을 사용함
  3. Stack만 별도 할당 = 독립적 || Code, Data, Heap 영역은 공유 = 메모리 공유(변수 공유)
  4. 한 thread의 결과가 다른 thread에 영향을 끼침
     → 동기화 문제 반드시 주의 (디버깅 어려움)

* Multi Thread
  1. 한 개의 단일 애플리케이션(응용프로그램) → 여러 thread로 구성 후 작업 처리
  2. 시스템 자원 소모 감소 → 효율성 증가 || 처리량 증가 → Cost 감소
  3. 통신 부담 감소 || 디버깅 어려움 || 단일 process에는 효과 미약 || 자원 공유 문제(교착상태=deadlock) || process에 영향을 줌

* Multi Process
  1. 한 개의 단일 애플리케이션(응용프로그램) → 여러 process로 구성 후 작업 처리
  2. 한 개의 프로세스 문제 발생은 확산 없음(process kill)
    - multithread는 하나의 thread 때문에 교착 상태가 일어나면 프로그램이 실행되는 건지 아닌지 알 수가 없음
  3. 캐시 체인지 주로 활용 || Cost 비용 매우 높은(오버헤드=overhead) || 복잡한 통신 방식 사용 (Process의 5. 설명 참고)

### 2) Python's GIL<br><br>

image.png
<br><br>

* GIL(Global Interpreter Lock)
  1. CPython → Python(bytecode) 실행 시 여러 thread를 사용할 경우 단일 thread만이 python object에 접근할 수 있도록 제한하는 mutex
    - CPython이 내가 작성한 .py 파일을 해석하여 Python(byte code)로 바꾸는 것이 python의 실행 원리 ▶ 싱글thread일 때가 python 언어의 실행 속도가 빨라서, 결국 GIL이라는 것을 만듦. 한 타임에 한 thread가 접근해서 I/O Bound ~?
  2. CPython 메모리 관리가 취약하기 때문 (즉, Thread-safe를 위함)
  3. 단일 thread로도 충분히 빠름
  4. process 사용 가능 = process는 GIL이 없으므로 multiprocess를 사용하는 작업을 python으로 만들어서 사용하면
    - ex) numpy, scipy 등 - multiprocess, 즉, GIL 외부 영역에서 효율적인 코딩이 가능하도록 라이브러리, 패키지들이 나와 있음.
  5. 병렬 처리는 multiprocessing, asyncio 등을 사용하면 됨 (선택지 다양)
  6. thread 동시성 완벽 처리를 위해서는 Jython, IronPython, Stackless Python 등을 사용하면 됨
